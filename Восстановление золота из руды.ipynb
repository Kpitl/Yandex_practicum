### –ù–∞—Ö–æ–∂–¥–µ–Ω–∏–µ MAE –º–µ–∂–¥—É –ø–æ–ª—É—á–µ–Ω–Ω—ã–º–∏ —Ä–∞—Å—á—ë—Ç–∞–º–∏ –∏ –∑–Ω–∞—á–µ–Ω–∏–µ–º –ø—Ä–∏–∑–Ω–∞–∫–∞

–î–ª—è —ç—Ç–æ–≥–æ –≤—ã—á–∏—Å–ª–∏–º –µ—ë –Ω–∞ –æ–±—É—á–∞—é—â–µ–π –≤—ã–±–æ—Ä–∫–µ –¥–ª—è –ø—Ä–∏–∑–Ω–∞–∫–∞ rougher.output.recovery. –ù–∞–π–¥–µ–º MAE –º–µ–∂–¥—É –ø–æ–ª—É—á–µ–Ω–Ω—ã–º–∏ –∑–Ω–∞—á–µ–Ω–∏—è–º–∏ –∏ –∑–Ω–∞—á–µ–Ω–∏–µ–º –ø—Ä–∏–∑–Ω–∞–∫–∞.

***–£—Å–ª–æ–≤–Ω—ã–µ –æ–±–æ–∑–Ω–∞—á–µ–Ω–∏—è:***
* C ‚Äî –¥–æ–ª—è –∑–æ–ª–æ—Ç–∞ –≤ –∫–æ–Ω—Ü–µ–Ω—Ç—Ä–∞—Ç–µ –ø–æ—Å–ª–µ —Ñ–ª–æ—Ç–∞—Ü–∏–∏/–æ—á–∏—Å—Ç–∫–∏ (rougher.output.concentrate_au);
* F ‚Äî –¥–æ–ª—è –∑–æ–ª–æ—Ç–∞ –≤ —Å—ã—Ä—å–µ/–∫–æ–Ω—Ü–µ–Ω—Ç—Ä–∞—Ç–µ –¥–æ —Ñ–ª–æ—Ç–∞—Ü–∏–∏/–æ—á–∏—Å—Ç–∫–∏ (rougher.input.feed_au);
* T ‚Äî –¥–æ–ª—è –∑–æ–ª–æ—Ç–∞ –≤ –æ—Ç–≤–∞–ª—å–Ω—ã—Ö —Ö–≤–æ—Å—Ç–∞—Ö –ø–æ—Å–ª–µ —Ñ–ª–æ—Ç–∞—Ü–∏–∏/–æ—á–∏—Å—Ç–∫–∏ (rougher.output.tail_au).

```python
# –ü—Ä–æ–∏–∑–≤–µ–¥–µ–º —Ä–∞—Å—á–µ—Ç —ç—Ñ—Ñ–µ–∫—Ç–∏–≤–Ω–æ—Å—Ç–∏ –æ–±–æ–≥–æ—â–µ–Ω–∏—è 
c = train_df['rougher.output.concentrate_au']
f = train_df['rougher.input.feed_au']
t = train_df['rougher.output.tail_au']
recovery = ((c*(f-t))/(f*(c-t))) * 100
print(recovery)
```

### –ò—Å—Å–ª–µ–¥–æ–≤–∞–Ω–∏–µ —Å—É–º–º–∞—Ä–Ω–æ–π –∫–æ–Ω—Ü–µ–Ω—Ç—Ä–∞—Ü–∏–∏ –≤—Å–µ—Ö –≤–µ—â–µ—Å—Ç–≤ –Ω–∞ —Ä–∞–∑–Ω—ã—Ö —Å—Ç–∞–¥–∏—è—Ö

–§–æ—Ä–º—É–ª–∞ –∏—Ç–æ–≥–æ–≤–æ–≥–æ SMAPE:


```python
# –ù–∞–ø–∏—à–∏–º —Ñ–æ—Ä–º—É–ª—É SMAPE —á–µ—Ä–µ–∑ —Ñ—É–Ω–∫—Ü–∏—é def
def calculation_smape(x, y):
    smape = (1/len(x))*np.sum((np.abs(x - y))/( (np.abs(x)+np.abs(y))/2))*100
    return smape
```

```python
# –ù–∞–ø–∏—à–µ–º —Ñ–æ—Ä–º—É–ª—É –∏—Ç–æ–≥–æ–≤–æ–≥–æ SMAPE
def final_smape_function(rougher, final):
    final = 0.25*rougher+0.75*final
    return final
```

```python
%%time
best_model_rf = 100
best_param_rf = 0
for n in range(10, 30, 20):
    for d in range(2,10,2):
        model_rf_rougher = RandomForestRegressor(n_estimators=n, max_depth=d, random_state=123)
        smape_rougher = cross_val_score(model_rf_rougher,
                                  features_train,
                                  target_rougher_train,
                                  scoring = smape_scorer,
                                  cv=5)
        final_smape_rougher = smape_rougher.mean()

        model_rf_final = RandomForestRegressor(n_estimators=n, max_depth=d, random_state=123)
        smape_final = cross_val_score(model_rf_final,
                                  features_train,
                                  target_final_train,
                                  scoring = smape_scorer,
                                  cv=5)
        final_smape_final = smape_final.mean()
        result_smape = final_smape_function(final_smape_rougher, final_smape_final)*(-1)
        if result_smape < best_model_rf:
            best_model_rf = result_smape
            best_param_rf = [n,d]
            
print('–†–µ–∑—É–ª—å—Ç–∞—Ç —Å–ª—É—á–∞–π–Ω–æ–≥–æ –ª–µ—Å–∞:', best_model_rf)
print('–õ—É—á—à–∏–µ –ø–∞—Ä–∞–º–µ—Ç—Ä—ã:', best_param_rf)
```
